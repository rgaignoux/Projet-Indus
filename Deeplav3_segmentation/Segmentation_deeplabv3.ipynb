```python
import torch
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os
from torchvision import transforms

def overlay_mask(original_image, binary_mask, color=[255, 0, 0], alpha=0.5):
    """
    Superpose un masque binaire coloré sur l'image originale sans assombrir les zones.
    """
    binary_mask_resized = Image.fromarray(binary_mask * 255).resize(original_image.size) #cette ligne utilise l'interpolation bilinéaire pour agrandir la catre de segmentation
    original_array = np.array(original_image)
    mask_array = np.array(binary_mask_resized) / 255.0

    overlay = np.zeros_like(original_array, dtype=np.uint8)
    overlay[:] = color

    blended = original_array.copy()
    blended[mask_array > 0] = (
        (1 - alpha) * blended[mask_array > 0] + alpha * overlay[mask_array > 0]
    ).astype(np.uint8)

    return Image.fromarray(blended)
def preprocess_image(image_path):
    """
    Charge et prétraite une image.
    """
    image = Image.open(image_path).convert("RGB")
    transform = transforms.Compose([
        transforms.Resize((1024, 1024)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    return transform(image).unsqueeze(0), image

def load_model(weights_path):
    """
    Charge un modèle pré-entraîné.
    """
    model = torch.load(weights_path, map_location=torch.device("cpu"))
    model.eval()
    print("Modèle chargé avec succès.")
    return model

def perform_segmentation(folder_path, model, threshold=0.7,save_dir="segmentation", color=[255, 0, 0]):
    """
    Effectue la segmentation sur toutes les images de la base de test.
    """
    os.makedirs(save_dir, exist_ok=True)

    for filename in os.listdir(folder_path):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(folder_path, filename)
            image_tensor, original_image = preprocess_image(image_path)

            with torch.no_grad():

                output = model(image_tensor)
                output_mask = torch.sigmoid(output[0][0]).numpy()

                # Redimensionnement pour correspondre à la taille d'entrée
                from torchvision.transforms.functional import resize
                output_resized = resize(torch.tensor(output_mask).unsqueeze(0).unsqueeze(0),
                                        size=original_image.size,
                                        interpolation=Image.BILINEAR)
                output_resized = output_resized.squeeze().numpy()  # Taille ajustée



                print(f"Taille de l'image originale : {original_image.size}")
                print(f"Taille brute de la carte de segmentation : {output_mask.shape}")
                print(f"Taille après redimensionnement : {output_resized.shape}")

                # Création du masque binaire
                binary_mask = (output_resized > threshold).astype(np.uint8)
                inverted_mask = 1 - binary_mask

                # Superposition sur l'image originale
                blended_image = overlay_mask(original_image, binary_mask, color=color)

                # Sauvegarde du masque inversé
                mask_image = Image.fromarray((inverted_mask * 255).astype(np.uint8))
                mask_path = os.path.join(save_dir, f"{os.path.splitext(filename)[0]}_mask.png")
                mask_image.save(mask_path)
                print(f"Masque inversé sauvegardé : {mask_path}")

                # Création de la carte de probabilité
                plt.figure(figsize=(6, 6))
                plt.imshow(output_mask, cmap='jet', alpha=0.6)
                plt.colorbar()
                plt.axis("off")
                plt.title("Carte de probabilité de segmentation")
                plt.savefig("temp_prob_map.png")
                plt.show()
                plt.close()

                # Chargement de la carte de probabilité
                prob_map = Image.open("temp_prob_map.png").resize(blended_image.size)
                prob_map_path = os.path.join(save_dir, f"{os.path.splitext(filename)[0]}_prob_map.png")
                prob_map.save(prob_map_path)
                print(f"Carte de probabilité sauvegardée : {prob_map_path}")



                segmented_image_path = os.path.join(save_dir, f"{os.path.splitext(filename)[0]}_segmented.png")
                blended_image.save( segmented_image_path)
                print(f"Image segmentée sauvegardée : {save_dir}")

weights_path = path+"/best_model.pth"
folder_path = path+"/DATABASE" # chemin vers le dataset
save_dir = path+"/image_test" #chemein vers le dossier de sauvegarde
model = load_model(weights_path)

perform_segmentation(folder_path, model, threshold=0.7,save_dir=save_dir,color=[0, 255, 0])
```

```python
import os
from PIL import Image
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np

def match_files(ground_truth_dir, predicted_masks_dir):
    """
    Crée une correspondance entre les fichiers de vérité terrain(axe central) et les fichiers prédits
    en utilisant les numéros présents dans leurs noms.
    """
    matches = []
    ground_truth_files = os.listdir(ground_truth_dir)
    predicted_files = os.listdir(predicted_masks_dir)

    for gt_file in ground_truth_files:
        gt_number = gt_file.split("_")[1]
        matching_pred_file = next(
            (pred for pred in predicted_files if gt_number in pred), None
        )
        if matching_pred_file:
            matches.append((gt_file, matching_pred_file))
        else:
            print(f"Aucun masque prédit trouvé pour {gt_file}")

    return matches

def calculate_metrics_with_matching_files(ground_truth_dir, predicted_masks_dir, threshold=0.7):
    """
    Calcule les métriques moyennes (Precision, Recall, F1 Score) .
    """
    precision_list = []
    recall_list = []
    f1_list = []

    matches = match_files(ground_truth_dir, predicted_masks_dir)

    for gt_file, pred_file in matches:
        ground_truth_path = os.path.join(ground_truth_dir, gt_file)
        predicted_mask_path = os.path.join(predicted_masks_dir, pred_file)

        # Charge les masques
        ground_truth = np.array(Image.open(ground_truth_path).convert("L")) / 255.0
        predicted_mask = np.array(Image.open(predicted_mask_path).convert("L")) / 255.0

        # Redimensionne le masque prédit pour correspondre au masque de vérité terrain
        predicted_mask = np.array(Image.fromarray(predicted_mask).resize(ground_truth.shape, Image.NEAREST))

        # Binarise les masques
        ground_truth_binary = (ground_truth > threshold).astype(np.uint8)
        predicted_mask_binary = (predicted_mask > threshold).astype(np.uint8)

        # Calcule les métriques
        precision = precision_score(ground_truth_binary.flatten(), predicted_mask_binary.flatten())
        recall = recall_score(ground_truth_binary.flatten(), predicted_mask_binary.flatten())
        f1 = f1_score(ground_truth_binary.flatten(), predicted_mask_binary.flatten())

        precision_list.append(precision)
        recall_list.append(recall)
        f1_list.append(f1)

        print(f"{gt_file} & {pred_file} -> Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}")

    # Moyenne des métriques
    if precision_list:
        avg_precision = np.mean(precision_list)
        avg_recall = np.mean(recall_list)
        avg_f1 = np.mean(f1_list)

        print("\n=== Résultats moyens ===")
        print(f"Average Precision: {avg_precision:.4f}")
        print(f"Average Recall: {avg_recall:.4f}")
        print(f"Average F1 Score: {avg_f1:.4f}")
    else:
        print("\nAucune métrique calculée (aucune correspondance).")
        avg_precision, avg_recall, avg_f1 = None, None, None


ground_truth_dir = "/content/Mydrive/MyDrive/axe_file"  # Chemin vers les axes centraux (verité terrain)
predicted_masks_dir = "/content/Mydrive/MyDrive/masks_file"  # Chemin vers les masques prédits
calculate_metrics_with_matching_files(ground_truth_dir, predicted_masks_dir, threshold=0.7)
```
